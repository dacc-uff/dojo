```python
from itertools import combinations, chain

def thug(weights):
    weight_set = set(range(len(weights)))
    return list(sorted(min([
        (sum(weights[y] for y in x),
         sum(weights[y] for y in (weight_set - set(x)))) 
        for x in chain.from_iterable(
            combinations(weight_set, x)
            for x in range(len(weights) + 1)
        )], key=lambda x: abs(x[0] - x[1])
    ), reverse=True))
    
    if len(weights) == 1:
        return [weights[0], 0]
    if len(weights) > 2:
        solutions=[
            [weights[0]+weights[1], weights[2]],
            [weights[0]+weights[2], weights[1]],
            [weights[1]+weights[2], weights[0]],
            
        ]
        return list(sorted(min(
            solutions,
            key=lambda x: abs(x[0] - x[1])
        ), reverse=True))
          
    if weights[0]<weights[1]:
        return list(reversed(weights))
    return weights
```

```python
def thug2 (lista):
    lista.sort(reverse = True)
    a = 0
    b = 0
    for i in range(len(lista)):
        if (a <= b):
            a += lista[i]
        else:
            b += lista[i]
    return sorted([a, b], reverse = True)
        

def thug3(lista):
    lista.sort(reverse=True)
    s = [0, 0]
    for elemento in lista:
        s[int(s[0] > s[1])] += elemento
    return sorted(s, reverse=True)
        
```
